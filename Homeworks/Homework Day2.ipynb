Create a list and swap the second half of the list with the first half of list and print this list on the screen
First Question but ask to user to number for change

```python
UserNumber = int(input("Write an even number:"))
List=[]
for i in range(1,UserNumber+2):
  List.append(i)
LenList = len(List)
HalfList = int(len(List)/2)
HalfChange = []

for j in range(HalfList+1,LenList):
  HalfChange.append(j)
for k in range(1,HalfList+1):
  HalfChange.append(k)
print(HalfChange)
```
